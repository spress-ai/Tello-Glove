```python
def to_array(df):
    x = []
    new_x = []
    for i in range(len(df.columns)):    
        x.append(df[str(i + 1)].to_numpy())

    for x_arr in x:
        x_arr_parsed = []
        for x_val in x_arr:
            x_arr_parsed.append(extract_array(x_val))
        new_x.append(x_arr_parsed)
        
    new_x = np.array(new_x)
    return np.array(new_x)
```

```python
def extract_array(string_array):
    arr = string_array.replace('\'', '').replace('[', '').replace(']', '').replace(' ', '').split(',')
    arr = map(lambda a : float(a), arr) 
    return list(arr)
```

```python
left_circle_x = to_array(left_circle_df)
left_circle_y = np.ones(len(left_circle_x))

random_movements_x = to_array(random_movements_df)
random_movements_y = np.zeros(len(random_movements_x))

x = np.concatenate((left_circle_x, random_movements_x), axis = 0)
y = np.concatenate((left_circle_y, random_movements_y), axis = 0)
#x = x[:, :, 0]#.reshape(40, 150, 1)
x = x[:, :, [3, 4, 5]]
```

```python
fwhm = 5
k = 8
gtime = np.arange(-k, k)

gauswin = np.exp(-(4*np.log(2) * gtime ** 2) / fwhm ** 2)
gauswin = gauswin / np.sum(gauswin)

def gaussian_filter(data):
    filtered_data = copy.deepcopy(data)
    for i in range(k + 1, data.shape[0] - k - 1):
        for j in range(data.shape[1]):
            filtered_data[i][j] = np.sum(data[i - k : i + k, j] * gauswin)
       
    return filtered_data
```

```python
def is_spike_up(threshold_up, threshold_down, x_window):
    spike = [None, None]

    for i in range(len(x_window) - 1):
        if x_window[i + 1] > x_window[i] * threshold_up:
            spike[0] = i
            
    if spike[0] != None:
        for i in range(spike[0], len(x_window) - 1):
            if x_window[i + 1] < x_window[i] * threshold_down:
                spike[1] = i
            
    return spike
            
def is_spike_down(threshold_up, threshold_down, x_window):
    spike = [None, None]

    for i in range(len(x_window) - 1):
        if x_window[i + 1] < x_window[i] * threshold_down:
            spike[0] = i
            
    if spike[0] != None:
        for i in range(spike[0], len(x_window) - 1):
            if x_window[i + 1] > x_window[i] * threshold_up:
                spike[1] = i
            
    return spike
        
def get_is_spike(x, x_window):
    threshold_up = 50
    threshold_down = 0.02
    is_spike = False
    spike = [None, None]
    
    spike_up = is_spike_up(threshold_up, threshold_down, x_window)
    spike_down = is_spike_down(threshold_up, threshold_down, x_window)
    
    if spike_up[0] != None and spike_up[1] != None:
        is_spike = True
        spike = spike_up
    elif spike_down[0] != None and spike_down[1] != None:
        is_spike = True
        spike = spike_down
        
    return (is_spike, spike)


def get_filterd_data(x):
    #print(x.shape)
    x = np.array(x)
    for i in range(len(x)):
        if i > 1 and i < len(x) - 2:
            for j in range(len(x[i])):
                x_window = x[i - 2 : i + 3][:, j] # need to be odd
                is_spike, spike_window_indexes = get_is_spike(x[i][j], x_window)
                if is_spike:
                    #print(spike_window_indexes)
                    #spikes_indexes = np.array([i - (len(x_window) - 1)/2 + spike_window_indexes[0], i - (len(x_window) - 1)/2 + spike_window_indexes[1]])
                    #new_value = (x[int(spikes_indexes[0] - 1)][j] + x[int(spikes_indexes[1] + 1)][j] )/2
                    #print(i, spike_window_indexes, spikes_indexes, new_value)
                    x[i][j] = np.median(x_window)
        #if not (x_val[1] == -90.00 or x_val[2] == 90.00 or x_val[10] > 1000 or x_val[10] < -1000 or x_val[11] > 1000 or  x_val[11] < -1000):
        #        filtered_x.append(x_val)
    return x
```

```python
def get_model():
    accX_input = Input(shape=(100, 1))
    accY_input = Input(shape=(100, 1))
    accZ_input = Input(shape=(100, 1))
    
    accX_CNN = Conv1D(filters = 64, kernel_size = 3, activation = 'relu')(accX_input)
    accY_CNN = Conv1D(filters = 64, kernel_size = 3, activation = 'relu')(accY_input)
    accZ_CNN = Conv1D(filters = 64, kernel_size = 3, activation = 'relu')(accZ_input)
    
    #accX_CNN = Dropout(0.1)(accX_CNN)
    #accY_CNN = Dropout(0.1)(accY_CNN)
    #accZ_CNN = Dropout(0.1)(accZ_CNN)
    
    accX_flatten = Flatten()(accX_CNN)
    accY_flatten = Flatten()(accY_CNN)
    accZ_flatten = Flatten()(accZ_CNN)
    
    accX_model = Model(inputs = accX_input, outputs = accX_flatten) 
    accY_model = Model(inputs = accY_input, outputs = accY_flatten) 
    accZ_model = Model(inputs = accZ_input, outputs = accZ_flatten) 
    
    combined = concatenate([accX_model.output, accY_model.output, accZ_model.output])
    dense_layer = Dense(100, activation = 'relu')(combined)
    dense_layer = Dense(30, activation = 'relu')(combined)
    #dense_layer = Dropout(0.5)(dense_layer)
    
    output = Dense(1, activation='sigmoid')(dense_layer)
    
    model = Model(inputs = [accX_model.input, accY_model.input, accZ_model.input], outputs = output)
    model.compile(optimizer = Adam(learning_rate = 0.001),
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])

    #model.add(Conv1D(filters = 32, kernel_size = 3, activation='relu'))
    #model.add(Dropout(0.2))
    return model
```

```python
def extract_array_by_sensors(x):
    to_return = []
    for i in range(x.shape[2]):
        acc_x = x[:, :, i]
   
        to_return.append(acc_x)
    

    '''for i in range(x.shape[0]):
        acc_x = [x[i, :, 0], x[i, :, 1], x[i, :, 2]]
   
        to_return.append(acc_x)
    '''
    
    return np.array(to_return)  
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=40)
x_train = x_train.reshape(x_train.shape[0], x_train.shape[1], 3)
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1], 3)

scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train.reshape(-1, x_train.shape[-1])).reshape(x_train.shape)
x_test = scaler.transform(x_test.reshape(-1, x_test.shape[-1])).reshape(x_test.shape)

#x_train = x_train.reshape(x_train.shape[0], x_train.shape[2], x_train.shape[1], 1)
#x_test = x_test.reshape(x_test.shape[0], x_test.shape[2], x_test.shape[1], 1)

x_train = extract_array_by_sensors(x_train)
x_test = extract_array_by_sensors(x_test)
x_train = x_train.reshape(x_train.shape[0], x_train.shape[1], x_train.shape[2], 1)
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1], x_test.shape[2], 1)

print(x_train.shape)
model = get_model()
model.fit(x = [x_train[0], x_train[1], x_train[2]], y = y_train, epochs = 10, validation_data=([x_test[0], x_test[1], x_test[2]], y_test), verbose=1)
```

```python
pred = model.predict([x_test[0], x_test[1], x_test[2]])
print(len(pred))
pred = pred > 0.5
pred = np.array(pred).reshape(20)
print(pred, y_test, np.sum(pred == y_test)/20*100)
```
